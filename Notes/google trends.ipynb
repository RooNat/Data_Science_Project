```python
from pytrends.request import TrendReq
import pandas as pd
import plotly.express as px
```

```python
pytrend=TrendReq(retries=3)
```

```python
kw_list = ['Bitcoin']
pytrend.build_payload(kw_list=kw_list, timeframe='2022-01-01 2022-05-06')

df = pytrend.interest_over_time()
df = df.reset_index()
# fig = px.line(df, x ='date', y = 'Bitcoin')
# fig.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df['date'], df['Bitcoin'],marker='o',linestyle='-' )
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Google Trends')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.to_csv("Bitcoin_google_2.csv")
```
